```python
# Imorting Modules

import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Reading File

df = pd.read_excel('flight_price.xlsx')

df.head()
```

```python
# Droping Route cause it is of no use

df.drop('Route',axis=1,inplace=True)
```

```python
# Date_of_Journey

df['Day'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

# Droping Date_of_Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
# Converting Day, Month, Year columns from str to int

df['Day']=df['Day'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)

df.info()
```

```python
df.head()
```

```python
# Dep_Time

df['Dep_Hours'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes']= df['Dep_Time'].str.split(':').str[1]

# Droping Dep_Time Column
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
# Converting Dep_Hours, Dep_Minutes columns from str to int
df['Dep_Hours'] = df['Dep_Hours'].astype(int)
df['Dep_Minutes'] = df['Dep_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
# Arrival_Time

df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0]) 
df['Arrival_Hours'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
#Dropping Arrival_Time Column

df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
#Converting Arrival_Hours, Arrival_Minutes columns from str to int

df['Arrival_Hours']= df["Arrival_Hours"].astype(int)
df['Arrival_Minutes']= df["Arrival_Minutes"].astype(int)
df.info()
```

```python
df.head()
```

```python
# Duration

df['Duration_Hours'] = df['Duration'].str.split(' ').str[0]
df['Duration_Minutes'] = df['Duration'].str.split(' ').str[1]

df['Duration_Hours'] = df['Duration_Hours'].apply(lambda x:x.split('h')[0])
df['Duration_Minutes'] = df['Duration_Minutes'].str.split('m').str[0]
```

```python
# Dropping Duration Column
df.drop('Duration',axis=1,inplace=True)
```

```python
df['Duration_Hours'].unique()
df['Duration_Hours'] = df['Duration_Hours'].str.extract('(\d+)').astype(int) #pandas searches each value and extracts only the numeric part.
```

```python
df['Duration_Minutes'] = df['Duration_Minutes'].fillna(0)
```

```python
# converting Duration_Hours, Duration_Minutes fron object to int
df['Duration_Hours'] = df['Duration_Hours'].astype(int) 
df['Duration_Minutes'] = df['Duration_Minutes'].astype(int)

```

```python
df.info()
```

```python
df.head()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3 ,'4 stops':4, np.nan:1})
```

```python
# converting Total_Stops column from object to int
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
# Now for Airline, Source, Destination, Additional_Info 
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray()
```

```python
ds = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
ds.head()
```

```python
df.drop(['Airline','Destination','Source','Additional_Info'], axis=1, inplace=True)
```

```python
final_data = pd.concat([ds.reset_index(drop=True),df.reset_index(drop=True)],axis=1)
```
